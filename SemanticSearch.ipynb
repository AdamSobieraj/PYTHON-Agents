```python
#function to generate embeddings 
def get_embedding(text, model="text-embedding-ada-002"):
   text = text.replace("\n", " ")
   return client.embeddings.create(input = [text], model=model).data[0].embedding
```

```python
#Create a database 
import numpy as np
import pandas as pd

# Embeddingi
embeddings = [get_embedding(text) for text in knowledgebase]

def robust_norm(x):
    x = np.array(x, dtype=float)

    # Pomiń NaN/inf
    if not np.isfinite(x).all():
        return None

    # Pomiń wektory zerowe
    if np.all(x == 0):
        return 0.0

    # Unikanie overflow
    max_val = np.max(np.abs(x))
    if max_val == 0:
        return 0.0
    scaled_x = x / max_val
    norm_scaled = np.sqrt(np.sum(scaled_x**2))
    return norm_scaled * max_val

cleaned_records = []

for text, vec in zip(knowledgebase, embeddings):

    norm = robust_norm(vec)

    # Normalizacja wektora
    vec = vec / norm
    cleaned_records.append((text, vec))

df = pd.DataFrame(cleaned_records, columns=["Text", "Vector"])
```

```python
import numpy as np
from scipy.spatial.distance import cosine, euclidean

# Wyłączamy ostrzeżenia numeryczne w NumPy (divide by zero, overflow itp.)
np.seterr(all='ignore')

def safe_vector(u, eps=1e-12, clip_value=1e100):
    u = np.array(u, dtype=np.float64)

    # Pomijamy NaN/inf
    if not np.isfinite(u).all():
        return None

    # Pomijamy wektory zerowe
    norm = np.linalg.norm(u)
    if norm < eps:
        return None

    # Ograniczamy ekstremalne wartości
    u = np.clip(u, -clip_value, clip_value)

    # Normalizujemy, żeby uniknąć overflow
    u /= norm
    return u

def safe_cosine(u, v):
    u_safe = safe_vector(u)
    v_safe = safe_vector(v)
    if u_safe is None or v_safe is None:
        return np.nan
    try:
        return cosine(u_safe, v_safe)
    except:
        return np.nan

def safe_euclidean(u, v):
    u_safe = safe_vector(u)
    v_safe = safe_vector(v)
    if u_safe is None or v_safe is None:
        return np.nan
    try:
        return euclidean(u_safe, v_safe)
    except:
        return np.nan

# Obliczenia w DataFrame
df['Cosine_Distance'] = df['Vector'].apply(lambda x: safe_cosine(x, query_vector))
df['Euclidean_Distance'] = df['Vector'].apply(lambda x: safe_euclidean(x, query_vector))
```

```python
#  Funkcje odległości
import numpy as np
from numpy.linalg import norm
import numpy as np

def safe_norm(x, eps=1e-12):
    x = np.array(x, dtype=np.float64)
    if not np.isfinite(x).all():
        return np.nan
    max_val = np.max(np.abs(x))
    if max_val < eps:
        return 0.0
    x_scaled = x / max_val
    return np.sqrt(np.sum(x_scaled**2)) * max_val

def safe_cosine_distance(a, b, eps=1e-12):
    a = np.array(a, dtype=np.float64)
    b = np.array(b, dtype=np.float64)
    if not np.isfinite(a).all() or not np.isfinite(b).all():
        return np.nan

    norm_a = safe_norm(a, eps)
    norm_b = safe_norm(b, eps)
    if norm_a < eps or norm_b < eps:
        return np.nan

    a_norm = a / norm_a
    b_norm = b / norm_b
    dot = np.sum(a_norm * b_norm)
    dot = np.clip(dot, -1.0, 1.0)
    return 1 - dot
```

```python
#  Wyszukiwanie najbardziej podobnych wpisów

def search_knowledgebase(query, top_x=5, distance="cosine"):
    query_vector = get_embedding(query)
    if distance == "cosine":
        df["Distance"] = df["Vector"].apply(lambda x: safe_cosine_distance(x, query_vector))
    else:
        df["Distance"] = df["Vector"].apply(lambda x: safe_euclidean_distance(x, query_vector))
    return df.nsmallest(top_x, "Distance")
```

```python
#  Generowanie odpowiedzi na podstawie kontekstu

def generate_answer(query, context_texts):
    context = "\n".join(context_texts)
    prompt = f"""
Jesteś asystentem sklepu z elektroniką.
Odpowiedz na pytanie klienta na podstawie poniższej wiedzy:

{context}

Pytanie: {query}

Odpowiedz pełnym zdaniem, po polsku, w sposób pomocny i naturalny.
"""
    response = client.chat.completions.create(
        model="gpt-4o-mini",
        messages=[{"role": "user", "content": prompt}]
    )
    return response.choices[0].message.content.strip()

# Główna logika – zapytanie użytkownika

query = input("Zadaj pytanie: ")

# Dynamiczny dobór x – więcej kontekstu dla krótkich pytań
x = 5 if len(query) < 50 else 3

top_matches = search_knowledgebase(query, top_x=x, distance="cosine")

print("\n--- Najbardziej dopasowane fragmenty ---")
for i, t in enumerate(top_matches["Text"].tolist(), start=1):
    print(f"{i}. {t}")

answer = generate_answer(query, top_matches["Text"].tolist())

print("\n Odpowiedź asystenta:\n")
print(answer)
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder

# Kodowanie kategorii
le = LabelEncoder()
y_train_enc = le.fit_transform(y_train)
y_test_enc = le.transform(y_test)

# KNN
knn = KNeighborsClassifier(n_neighbors=3, metric='cosine', n_jobs=-1)
knn.fit(X_train, y_train_enc)

# Predykcja
y_pred_enc = knn.predict(X_test)
y_pred = le.inverse_transform(y_pred_enc)
```
